```python
import pyglet
from pyglet.canvas import Display
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("../src/drums.jpg")
```

```python
pic = pyglet.image.ImageData(im.shape[1], im.shape[0], 'RGB', im.tobytes(), -3*im.shape[1])
```

```python
screens = Display().get_screens()
```

```python
screens
```

```python
window = pyglet.window.Window(fullscreen=True, screen=screens[1])
```

```python
@window.event
def on_draw():
    window.clear()
    pic.blit(0, 0)
    
@window.event
def on_key_press(symbol, modifier):
    # key "E" get press
    if symbol == pyglet.window.key.E:
        # close the window
        window.close()
```

```python
pyglet.app.run()
```
