```python
import sounddevice as sd
import soundfile as sf
import os
import numpy as np
import time
import matplotlib.pyplot as plt
import multiprocessing
```

```python
def play_sound():
    sound_dir = "/home/akash/Documents/muTable/src/instruments/drums/sound_data/"

    sound1 = os.path.join(sound_dir, "3.wav")
    sound2 = os.path.join(sound_dir, "1.wav")
    sound3 = os.path.join(sound_dir, "2.wav")

    # Extract data and sampling rate from file
    data1, fs1 = sf.read(sound1, dtype='float32')  
    data2, fs2 = sf.read(sound2, dtype='float32')  
    data3, fs3 = sf.read(sound3, dtype='float32')

    while(1):
        for i in range(2):
            sd.play(data1, fs1)
            time.sleep(0.3)
            sd.play(data2, fs2)
            time.sleep(0.3)
        for i in range(2):
            sd.play(data3, fs3)
            time.sleep(0.3)
            sd.play(data2, fs2)
            time.sleep(0.3)
```

```python
p = multiprocessing.Process(target=play_sound, args=())
```

```python
p.start()
```
