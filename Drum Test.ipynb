```python
import sounddevice as sd
import soundfile as sf
import os
import time
import matplotlib.pyplot as plt
```

```python
sound_dir = "./data/drum_sounds/"
```

```python
sound1 = os.path.join(sound_dir, "big.wav")
sound2 = os.path.join(sound_dir, "small.wav")
sound3 = os.path.join(sound_dir, "ring.wav")
```

```python
# Extract data and sampling rate from file
data1, fs1 = sf.read(sound1, dtype='float32')  
data2, fs2 = sf.read(sound2, dtype='float32')  
data3, fs3 = sf.read(sound3, dtype='float32')
```

```python
data1 = data1[:40000,:]
data1[:,0] = data1[:,0]/3

data2 = data2[:40000,:]
data2[:,1] = data2[:,1]/3

data3 = data3[:8000,:]
data3[:,0] = data3[:,0]/3
```

```python
plt.plot(data2[:60000,0])
```

```python
while(1):
    for i in range(2):
        sd.play(data1, fs1)
        time.sleep(0.3)
        sd.play(data2, fs2)
        time.sleep(0.3)
    for i in range(2):
        sd.play(data3, fs3)
        time.sleep(0.3)
        sd.play(data2, fs2)
        time.sleep(0.3)
    
```
