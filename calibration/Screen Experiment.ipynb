```python
import pyglet
from pyglet.canvas import Display
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("markers.jpg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
im = cv2.resize(im, (1080,1080), interpolation = cv2.INTER_AREA)
```

```python
im.shape
```

```python
plt.imshow(im)
```

```python
pic = pyglet.image.ImageData(im.shape[0], im.shape[1], 'RGB', im.tobytes(), -1*im.shape[1]*3)
```

```python
screens = Display().get_screens()
```

```python
screens
```

```python
window = pyglet.window.Window(fullscreen=True, screen=screens[1])
```

```python
@window.event
def on_draw():
    window.clear()
    pic.blit(0, 0)

pyglet.app.run()
```

```python
window.close()
```
